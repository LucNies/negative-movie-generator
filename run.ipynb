```python
import lstm
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
%matplotlib notebook
from keras import utils

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import CuDNNLSTM as LSTM
from keras.callbacks import ModelCheckpoint
```

```python
lstm.train()
```

```python
raw_text, chars, dataX, dataY, int_to_char = lstm.prepare_data()
```

```python
def make_model(input_shape, output_dim, batch_size = None, stateful=False, dropout=0.2):
    print("model dim: ", input_shape, output_dim)
    model = Sequential()
    model.add(LSTM(256, input_shape=input_shape, batch_size=batch_size, return_sequences=True, stateful=stateful))
    model.add(Dropout(dropout))
    model.add(LSTM(128,  return_sequences=True, stateful=stateful))
    model.add(Dropout(dropout))
    model.add(LSTM(64,  return_sequences=True, stateful=stateful))
    model.add(Dropout(dropout))
    model.add(LSTM(128,  return_sequences=True, stateful=stateful))
    model.add(Dropout(dropout))
    model.add(LSTM(256,  return_sequences=False, stateful=stateful))
    model.add(Dropout(dropout))
    
    model.add(Dense(output_dim, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam')
    return model

```

```python
#Make model
shape = (dataX.shape[1], dataX.shape[2])
model = make_model(shape, dataY.shape[1], batch_size=None)

filepath = os.path.join('models', "weights-improvement-{epoch:02d}-bigger.hdf5")
checkpoint = ModelCheckpoint(filepath, monitor='loss', verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint]
model.fit(dataX, dataY, epochs=50, batch_size=512, callbacks=callbacks_list)
```

```python
seq_length = 64
# load ascii text and covert to lowercase
filename = "wonderland.txt"
raw_text = open(filename, encoding="utf8").read()
raw_text = raw_text.lower()

# create mapping of unique chars to integers
chars = sorted(list(set(raw_text)))
char_to_int = dict((c, i) for i, c in enumerate(chars))
int_to_char = {number : character for (number, character) in zip(char_to_int.values(), char_to_int.keys())}

n_chars = len(raw_text)
n_vocab = len(chars)

print("Total Characters: ", n_chars)
# Convert text to integers and do one-hot encoding
int_text = [char_to_int[c] for c in raw_text]
int_text = utils.to_categorical(int_text)
dataX = []
dataY = []

#Split sequences
for i in range(0, n_chars - seq_length, 1):
    seq_in = int_text[i:i + seq_length]
    seq_out = int_text[i + seq_length]


    dataX.append(seq_in)
    dataY.append(seq_out)
    
n_patterns = len(dataX)
print("Total Patterns: ", n_patterns)
dataX = np.array(dataX)
dataY = np.array(dataY)

```

```python
# load the network weights
# For inference!
#shape = (dataX.shape[1], dataX.shape[2])
model = make_model((1, dataX.shape[2]), dataY.shape[1], batch_size=1)
filename = os.path.join('models', "weights-improvement-{}-bigger.hdf5".format(50))
model.load_weights(filename)
model.compile(loss='categorical_crossentropy', optimizer='adam')
```
